### **1. Introduction to Probability and Distributions**
#### **Concept**: Understanding data distributions is fundamental to making inferences. Probability helps quantify uncertainty, while distributions provide a way to understand how data behaves.

#### **Topics & Key Concepts**:
- **Probability Basics**: Concept of chance, probability range (0 to 1), and simple examples.
- **Distributions**:
  - **Normal Distribution**: Symmetrical, bell-shaped curve, common in biological data (e.g., human heights, lab measurements).
  - **Uniform Distribution**: Equal probability for all outcomes (e.g., rolling a fair die).

#### **Examples**:
1. **Normal Distribution in Biological Measurements**:
   - **Scenario**: Heights of a sample of individuals in a biology study.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a normal distribution (mean height 170 cm, std dev 10 cm)
heights = np.random.normal(loc=170, scale=10, size=500)

# Plot the distribution
plt.hist(heights, bins=30, color='lightblue', edgecolor='black')
plt.title("Height Distribution of Sample Population")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

**Explanation**: A histogram of heights, showing the normal distribution commonly found in biological traits.


2. **Uniform Distribution in Chemical Measurements**:
   - **Scenario**: Simulating the measurement error in a lab experiment.

```python
# Generate a uniform distribution between 0 and 1 (measurement error)
measurement_error = np.random.uniform(low=0, high=1, size=500)

# Plot the uniform distribution
plt.hist(measurement_error, bins=20, color='orange', edgecolor='black')
plt.title("Uniform Distribution of Measurement Error")
plt.xlabel("Error Value")
plt.ylabel("Frequency")
plt.show()
```

**Explanation**: A uniform distribution representing consistent error range in measurements.

### **2. Understanding Sampling and Sample Size**
#### **Concept**: Sampling allows data analysis on a smaller, manageable dataset. A representative sample size is critical for accurate conclusions.



#### **Topics & Key Concepts**:
- **Population vs. Sample**: Importance of a good sample to represent the entire population.
- **Sample Size Impact**: How different sample sizes affect statistical results.
- **Sampling Methods**: Simple random sampling, stratified sampling, etc.



#### **Examples**:
1. **Sampling Blood Pressure Measurements**:
   - **Scenario**: Measuring average blood pressure in a clinic.


```python
# Full dataset representing population blood pressure
blood_pressure = np.random.normal(loc=120, scale=15, size=1000)

# Sample 50 patients randomly
sample = np.random.choice(blood_pressure, size=50, replace=False)

plt.hist(blood_pressure, bins=30, alpha=0.5, label='Population', color='lightblue')
plt.hist(sample, bins=15, alpha=0.7, label='Sample', color='green')
plt.title("Blood Pressure: Population vs. Sample")
plt.xlabel("Blood Pressure (mmHg)")
plt.ylabel("Frequency")
plt.legend()
```

**Explanation**: A comparison of a full population distribution versus a sample distribution.

### **3. Introduction to Hypothesis Testing**
#### **Concept**: Hypothesis testing determines whether there is enough evidence in a sample of data to infer that a certain condition holds true for the entire population.

#### **Topics & Key Concepts**:
- **Null Hypothesis (H)**: Assumes no effect or difference.
- **Alternative Hypothesis (H)**: Assumes there is an effect or difference.
- **Significance Level ()**: Threshold for rejecting the null hypothesis (commonly set at 0.05).
- **p-value**: Probability of observing the results given that the null hypothesis is true.
- **Types of Tests**: Parametric (assumes normality), Non-parametric (does not assume normality).

#### **Examples**:
1. **Testing the Effect of a Drug on Blood Pressure**:
   - **Scenario**: Comparing blood pressure before and after administering a drug.

```python
from scipy import stats

# Blood pressure before and after treatment
before_treatment = np.random.normal(loc=140, scale=10, size=30)
after_treatment = np.random.normal(loc=130, scale=10, size=30)

# Perform paired t-test
t_stat, p_value = stats.ttest_rel(before_treatment, after_treatment)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
```

**Explanation**: A paired t-test evaluates if the mean blood pressure significantly dropped after treatment.

### **4. Concept of p-value and t-tests**
#### **Concept**: The p-value helps determine the strength of evidence against the null hypothesis. A t-test is used to compare the means of two groups.

#### **Topics & Key Concepts**:
- **p-value Interpretation**: A smaller p-value indicates stronger evidence against H.
- **t-test Variants**: 
  - **One-sample t-test**: Compare sample mean to a known value.
  - **Two-sample t-test**: Compare means of two independent samples.
  - **Paired t-test**: Compare means of two related samples (e.g., pre- and post-treatment).

#### **Examples**:
1. **One-sample t-test**:
   - **Scenario**: Testing if the average sodium level in a sample differs significantly from a known standard value (135 mmol/L).

```python
sodium_levels = np.random.normal(loc=137, scale=5, size=40)

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(sodium_levels, 135)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
```

**Explanation**: Checks if the sample mean sodium level is different from the standard level.
