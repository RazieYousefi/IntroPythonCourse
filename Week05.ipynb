### **1. Introduction to Matplotlib**
#### **Concept**: Matplotlib is a powerful Python library for creating static, animated, and interactive visualizations.

**Basic Setup**:
- Importing Matplotlib
- Setting up basic parameters for plots

```python
import matplotlib.pyplot as plt
```

#### **Examples**:
1. **Line Plots**:
   - Plot changes in body temperature over time for a group of patients.

```python
time = [0, 1, 2, 3, 4, 5]
temperature = [98.6, 98.8, 99.1, 100.3, 101.2, 99.9]

plt.plot(time, temperature, color='blue', marker='o')
plt.title("Patient Body Temperature Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (Â°F)")
plt.show()
```

2. **Bar Charts**:
   - Compare the average recovery time of patients receiving different treatments.

```python
treatments = ['Treatment A', 'Treatment B', 'Treatment C']
recovery_time = [5, 7, 4]

plt.bar(treatments, recovery_time, color='green')
plt.title("Average Recovery Time for Different Treatments")
plt.xlabel("Treatment")
plt.ylabel("Recovery Time (days)")
plt.show()
```

3. **Histograms**:
   - Plot a histogram of serum potassium levels in a patient group.

```python
potassium_levels = [4.0, 4.2, 4.5, 4.1, 3.9, 4.8, 4.3, 4.6, 5.1, 3.8, 2.6, 3.5, 3.2, 4.1, 4.3, 5.2, 5.4]

plt.hist(potassium_levels, bins=5, color='purple')
plt.title("Distribution of Serum Potassium Levels")
plt.xlabel("Potassium Level (mmol/L)")
plt.ylabel("Frequency")
plt.show()
```

### **2. Customizing Plots in Matplotlib**
#### **Concept**: Customizing labels, titles, colors, and figure sizes makes plots easier to interpret.

#### **Examples**:
1. **Adding Labels and Titles**:
   - Using the line plot of body temperature, add a title, axis labels, and a legend.

```python
plt.plot(time, temperature, color='blue', marker='o', label='Patient Temperature')
plt.title("Patient Body Temperature Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (Â°F)")
plt.legend()
plt.show()
```

2. **Changing Colors and Figure Size**:
   - Adjust the color scheme and figure size for clarity.

```python
plt.figure(figsize=(8, 5))
plt.bar(treatments, recovery_time, color='teal')
plt.title("Average Recovery Time for Different Treatments")
plt.xlabel("Treatment")
plt.ylabel("Recovery Time (days)")
plt.show()
```

3. **Adding Grid and Adjusting Axes**:
   - Customize axes range and add a grid to the potassium level histogram.

```python

plt.hist(potassium_levels, bins=5, color='purple')
plt.title("Distribution of Serum Potassium Levels")
plt.xlabel("Potassium Level (mmol/L)")
plt.ylabel("Frequency")
plt.grid(True)
plt.xlim(3.5, 5.5)
plt.ylim(0, 5)
plt.show()
```

### **3. Introduction to Seaborn for Advanced Visualizations**
#### **Concept**: Seaborn is a Python library built on top of Matplotlib, with simpler syntax and additional visualization options ideal for data analysis.

**Basic Setup**:
- Import Seaborn and load sample datasets.

```python
import seaborn as sns
```

#### **Examples**:
1. **Bar Plot with Seaborn**:
   - Create a bar plot showing the average heart rate by age group.

```python
age_group = ['18-30', '31-45', '46-60', '60+']
avg_heart_rate = [72, 75, 70, 65]

sns.barplot(x=age_group, y=avg_heart_rate, palette="viridis")
plt.title("Average Heart Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Heart Rate (bpm)")
plt.show()
```

2. **Box Plot**:
   - Show the distribution of cholesterol levels by age group using a box plot.

```python
import pandas as pd
```

```python
cholesterol_data = {
    "Age_Group": ['18-30', '31-45', '46-60', '60+', '18-30', '31-45', '46-60', '60+', '18-30', '31-45'],
    "Cholesterol_Level": [180, 195, 210, 190, 175, 200, 215, 180, 185, 198]
}

df = pd.DataFrame(cholesterol_data)
sns.boxplot(hue="Age_Group", y="Cholesterol_Level", data=df, palette="mako")
#sns.boxplot(x="Age_Group", y="Cholesterol_Level", data=df, palette="mako") # try this to see the difference.
plt.title("Cholesterol Levels by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Cholesterol Level (mg/dL)")
plt.show()
```

3. **Histogram with Kernel Density Estimate (KDE)**:
   - Plot a histogram with KDE of blood sugar levels to understand the distribution.

```python
blood_sugar_levels = [90, 95, 100, 105, 110, 90, 88, 97, 101, 99, 97, 102]

sns.histplot(blood_sugar_levels, kde=True, color='coral')
plt.title("Distribution of Blood Sugar Levels")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Frequency")
plt.show()
```
